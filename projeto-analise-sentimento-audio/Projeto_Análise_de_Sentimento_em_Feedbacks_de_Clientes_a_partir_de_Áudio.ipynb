```python
from google.colab import files
import whisper

# Fazer upload de vários arquivos
uploaded = files.upload()

# Carregar o modelo Whisper
model = whisper.load_model("base")

# Criar uma lista para armazenar as transcrições
transcriptions = []

# Processar cada arquivo carregado
for audio_file in uploaded.keys():
    print(f"Processando o arquivo: {audio_file}")

    # Transcrever o áudio
    result = model.transcribe(audio_file)

    # Salvar a transcrição na lista
    transcriptions.append(
        result["text"]
    )



```

```python
result_df = pd.DataFrame()


result_df['text'] = transcriptions
result_df['textblob_sentiment']=None
result_df['vader_sentiment'] =None
```

```python
from textblob import TextBlob


def analyze_textblob(text):
    analysis = TextBlob(text)
    polarity = analysis.sentiment.polarity
    if polarity > 0:
        return 'positivo'
    elif polarity == 0:
        return 'neutro'
    else:
        return 'negativo'

result_df['textblob_sentiment'] = result_df['text'].apply(analyze_textblob)
```

```python
result_df
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

vader_analyzer = SentimentIntensityAnalyzer()

def analyze_vader(text):
    score = vader_analyzer.polarity_scores(text)['compound']
    if score > 0:
        return 'positivo'
    elif score == 0:
        return 'neutro'
    else:
        return 'negativo'


result_df['vader_sentiment'] = result_df['text'].apply(analyze_vader)
```

```python
result_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


sentiment_counts = result_df[['textblob_sentiment', 'vader_sentiment']].apply(pd.Series.value_counts)

sentiment_counts.plot(kind='bar')
plt.title('Comparação de Sentimentos entre Bibliotecas')
plt.xlabel('Sentimento')
plt.ylabel('Contagem')
plt.show()
```
